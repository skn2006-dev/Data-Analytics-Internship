```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y1 = [1, 4, 9, 16, 25]
y2 = [2, 4, 6, 8, 10]
y3 = [1, 2, 3, 4, 5]

plt.plot(x, y1, label='Squared', linestyle=':', marker='.', color='blue')
plt.plot(x, y2, label='Linear', linestyle='--', marker='s', color='green')
plt.plot(x, y3, label='Linear', linestyle=':', marker='s', color='red')

plt.title('Line Plot Variation')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

labels = ['A', 'B', 'C']
values1 = [10, 20, 15]
values2 = [12, 18, 20]
values3 = [15, 25, 22]

x = np.arange(len(labels))
width = 0.20

plt.bar(x - width, values1, width, label='Group 1', color='skyblue')
plt.bar(x, values2, width, label='Group 2', color='orange')
plt.bar(x + width, values3, width, label='Group 3', color='green')

plt.xticks(x, labels)
plt.ylabel('Values')
plt.title('Grouped Bar Chart')
plt.legend()
plt.show()
```

```python
categories = ['X', 'Y', 'Z']
values = [7, 12, 5]

plt.barh(categories, values, color='teal')
plt.title('Horizontal Bar Chart')
plt.xlabel('Value')
plt.show()
```

```python
data = np.random.randn(1000)

plt.hist(data, bins=30, density=True, alpha=0.7, color='purple', edgecolor='black')
plt.title('Histogram with Density')
plt.xlabel('Data')
plt.ylabel('Density')
plt.grid(True)
plt.show()
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
colors = np.random.rand(50)
sizes = 100 * np.random.rand(50)

plt.scatter(x, y, c=colors, s=sizes, alpha=0.5, cmap='viridis')
plt.colorbar()
plt.title('Scatter Plot with Color and Size')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()

```

```python
labels = ['Python', 'Java', 'C++', 'Go']
sizes = [45, 30, 15, 10]
explode = (0.1, 0, 0, 0)

plt.pie(sizes, labels=labels, explode=explode, autopct='%1.1f%%', startangle=90, shadow=True)
plt.title('Pie Chart Example')
plt.axis('equal')
plt.show()

```

```python
data = [7, 15, 13, 12, 10, 15, 21, 22, 23, 19]

plt.boxplot(data)
plt.title('Boxplot Example')
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]

plt.step(x, y,  color='green')
plt.title('Step Plot')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python

days = [1, 2, 3, 4, 5]
sleep = [7, 8, 6, 11, 7]
eat = [2, 3, 4, 3, 2]
work = [7, 8, 7, 6, 8]
play = [8, 5, 7, 4, 7]

plt.stackplot(days, sleep, eat, work, play, labels=['Sleep', 'Eat', 'Work', 'Play'], colors=['#FF9999','#66B2FF','#99FF99','#FFCC99'])
plt.legend(loc='upper left')
plt.title('Stack Plot Example')
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]
errors = [0.5, 1, 0.75, 0.3, 1.2]

plt.errorbar(x, y, yerr=errors, fmt='-o', capsize=5)
plt.title('Error Bar Plot')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
theta = np.linspace(0, 2 * np.pi, 100)
r = np.abs(np.sin(theta))

ax = plt.subplot(projection='polar')
ax.plot(theta, r)
plt.title('Polar Plot')
plt.show()
```

```python
fig, ax1 = plt.subplots()

x = [0, 1, 2, 3]
y1 = [0, 1, 4, 9]
y2 = [0, 1, 2, 3]

ax2 = ax1.twinx()

ax1.plot(x, y1, 'g-', label='Y1')
ax2.plot(x, y2, 'b--', label='Y2')

ax1.set_xlabel('X Axis')
ax1.set_ylabel('Y1', color='g')
ax2.set_ylabel('Y2', color='b')
plt.title('Twin Y-Axis Plot')
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

x = np.linspace(-5, 5, 50)
y = np.linspace(-5, 5, 50)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))

ax.plot_surface(X, Y, Z, cmap='viridis')
ax.set_title('3D Surface Plot')
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('netflix_titles.csv')
df = df[df['type'] == 'Movie']
df['duration'] = df['duration'].str.replace(' min', '').astype(float)

sns.scatterplot(data=df, x='release_year', y='duration', alpha=0.5)
plt.title('Movie Duration vs Release Year')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('netflix_titles.csv')
counts = df['type'].value_counts()

plt.pie(counts, labels=counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Movies and TV Shows on Netflix')
plt.axis('equal')  
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('netflix_titles.csv')
top_ratings = df['rating'].value_counts().head(10)

top_ratings.plot(kind='bar', color='skyblue')
plt.title('Top 10 Content Ratings on Netflix')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
df_movies = df[df['type'] == 'Movie']
df_movies['duration'] = df_movies['duration'].str.replace(' min', '').astype(float)
df_movies['duration'].plot(kind='hist', bins=30, title='Distribution of Movie Durations')
plt.xlabel('Duration (minutes)')
plt.show()
```

```python
users = pd.DataFrame({
    'user_id': [1,2,3,4],
    'user_type': ['Free','Premium','Free','Premium']
})
activity=pd.DataFrame({
    'user_id':[1,2,1,3,4,2,4],
    'login_count':[5,15,7,8,10,20,15],
    'rating':[3.5,4.2,3.9,4.0,4.8,4.5,4.7]
})    
```

```python
activity.groupby("user_id").agg({
    "login_count":"sum",
    "rating":"mean",
})
```

```python
activity.merge(users,on="user_id")
```

```python
pd.merge(users,activity,on="user_id")
```

```python
pd.merge(users,activity,on="user_id",how="inner")
```

```python
df1=pd.DataFrame({
    'user_id':[1,2,3],
    'name':['Alice','Bob','Charlie']
})
df2=pd.DataFrame({
    'user_id':[2,3,4],
    'age':[25,30,22]
})
merged_df=pd.merge(df1,df2,on='user_id',how='inner')
merged_df
```

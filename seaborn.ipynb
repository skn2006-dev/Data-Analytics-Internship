```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")

sns.scatterplot(data=tips, x="total_bill", y="tip", hue="sex", style="smoker", palette="deep")
plt.title("Scatter Plot with Multiple Semantic Mappings")
plt.show()
```

```python
flights = sns.load_dataset("flights")

sns.lineplot(data=flights, x="year", y="passengers", hue="month", palette="tab10")
plt.title("Monthly Passengers Over Years")
plt.show()
```

```python
sns.relplot(data=tips, x="total_bill", y="tip", hue="smoker", col="time", kind="scatter")
plt.suptitle("Relational Plot Faceted by Time", y=1.02)
plt.show()
```

```python
sns.barplot(data=tips, x="day", y="total_bill", hue="sex", palette="pastel")
plt.title("Average Total Bill by Day and Sex")
plt.show()
```

```python
sns.countplot(data=tips, x="day", hue="sex", palette="muted")
plt.title("Count of Records by Day and Sex")
plt.show()
```

```python
sns.boxplot(data=tips, x="day", y="total_bill", hue="sex", palette="Set3")
plt.title("Box Plot of Total Bill by Day and Sex")
plt.show()
```

```python
sns.violinplot(data=tips, x="day", y="total_bill", hue="sex",  palette="Set2")
plt.title("Violin Plot of Total Bill by Day and Sex")
plt.show()
```

```python
sns.stripplot(data=tips, x="day", y="total_bill", hue="sex", dodge=True, jitter=True, palette="Set1")
plt.title("Strip Plot of Total Bill by Day and Sex")
plt.show()
```

```python
sns.swarmplot(data=tips, x="day", y="total_bill", hue="sex", dodge=True, palette="Set2")
plt.title("Swarm Plot of Total Bill by Day and Sex")
plt.show()
```

```python
sns.catplot(data=tips, x="day", y="total_bill", hue="sex", kind="bar", col="time",  palette="Set3")
plt.suptitle("Violin Plots of Total Bill by Day, Sex, and Time", y=1.02)
plt.show()
```

```python
sns.histplot(data=tips, x="total_bill", bins=20, kde=True, color="skyblue")
plt.title("Histogram of Total Bill")
plt.show()
```

```python
sns.kdeplot(data=tips, x="total_bill", hue="sex", fill=True, common_norm=False, palette="crest", alpha=0.5)
plt.title("KDE Plot of Total Bill by Sex")
plt.show()
```

```python
sns.displot(data=tips, x="total_bill", col="sex", kde=True, bins=20, color="purple")
plt.suptitle("Distribution of Total Bill by Sex", y=1.02)
plt.show()
```

```python
sns.regplot(data=tips, x="total_bill", y="tip", scatter_kws={"color": "blue"}, line_kws={"color": "red"})
plt.title("Regression Plot of Tip vs. Total Bill")
plt.show()
```

```python
sns.lmplot(data=tips, x="total_bill", y="tip", hue="smoker", col="sex", height=5, aspect=1)
plt.suptitle("Linear Model Plots of Tip vs. Total Bill by Sex and Smoker", y=1.02)
plt.show()
```

```python
sns.residplot(data=tips, x="total_bill", y="tip", color="g")
plt.title("Residual Plot of Tip vs. Total Bill")
plt.show()
```

```python
import numpy as np
import pandas as pd

data = np.random.rand(10, 10)
data_df = pd.DataFrame(data, columns=[f'Col_{i}' for i in range(1, 11)], index=[f'Row_{i}' for i in range(1, 11)])
sns.heatmap(data_df, annot=True, fmt=".2f", cmap="viridis", linewidths=0.5)
plt.title("Heatmap of Random Data")
plt.show()
```

```python
flights_pivot = flights.pivot(index="month", columns="year", values="passengers")
sns.clustermap(flights_pivot, cmap="coolwarm", standard_scale=1)
plt.title("Cluster Map of Monthly Passengers")
plt.show()
```

```python
penguins = sns.load_dataset("penguins")

sns.pairplot(penguins, hue="species", palette="husl")
plt.suptitle("Pair Plot of Penguin Measurements", y=1.02)
plt.show()
```

```python
sns.jointplot(data=penguins, x="flipper_length_mm", y="bill_length_mm", hue="species", kind="kde", fill=True)
plt.suptitle("Joint KDE Plot of Flipper Length vs. Bill Length", y=1.02)
plt.show()
```

```python
g = sns.FacetGrid(tips, col="sex", row="smoker", margin_titles=True)
g.map(sns.scatterplot, "total_bill", "tip")
g.fig.subplots_adjust(top=0.9)
g.fig.suptitle("Facet Grid of Total Bill vs. Tip by Sex and Smoker")
plt.show()
```
